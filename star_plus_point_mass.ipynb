```python
#MASS FRACTION-EPSILON NUC
#%pylab
import mesa_reader as mr
import numpy as np
import matplotlib.pyplot as plt

# load entire LOG directory information
l_1 = mr.MesaLogDir('LOGS1-2.2')
# grab the last profile
p_1 = l_1.profile_data(profile_number=29)

# this works even if you only have logRho and logT!
fig, ax1 = plt.subplots()
ax1.semilogy(p_1.R, p_1.h1, label='H$^1$')
ax1.semilogy(p_1.R, p_1.he4, label='He$^4$')
ax1.semilogy(p_1.R, p_1.c12, label='C$^{12}$')
ax1.tick_params(axis='y')

ax1.set_xlim(0,0.1)
ax1.set_ylim(0.000004,100)
ax1.set_xlabel("R (R$_\odot$)")
ax1.set_ylabel("Mass Fraction")
plt.legend()

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

ax2.set_ylabel('$\epsilon_{nuc}\ (erg\ g^{-1}\ s^{-1})$')  # we already handled the x-label with ax1
ax2.plot(p_1.R, p_1.eps_nuc, label='0.47 M$_\odot$', linestyle='dashed', color='black')
ax2.tick_params(axis='y')
plt.text(0.035, 10**5, r'$\epsilon_{nuc}$', fontsize = 12)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()

#plt.savefig('mass_frac-epsilon.jpeg', dpi=1200)
```

# Evolution

```python
#center-abundance-evolution

# make a MesaData object from a history file
h = mr.MesaData('history.data')

# extract the star_age column of data
ages = h.data('star_age')
h1 = h.data('center_h1')
he4 = h.data('center_he4')
c12 = h.data('center_c12')
o16 = h.data('center_o16')

# or do it more succinctly
ages = h.star_age

plt.semilogy(ages / 10**9, h1, label='H$^1$')
plt.semilogy(ages / 10**9, he4, label='He$^4$')
plt.semilogy(ages / 10**9, c12, label='C$^{12}$')
plt.semilogy(ages / 10**9, o16, label='O$^{16}$')

plt.xlabel("Age (Gyr)")
plt.ylabel("Center Abundance")
plt.xlim(0,10)
plt.ylim(10**-6,10)
plt.legend()
```

```python
#Kippenhahn-diagram-1

# make a MesaData object from a history file
h = mr.MesaData('history.data')

# extract the star_age column of data
ages = h.data('star_age')
log_Teff = h.data('log_Teff')
mass = h.data('star_mass')
luminosity = h.data('log_L')

plt.plot(ages / 10**9, log_Teff, color='black', label='log(T$_e)$')
plt.plot(ages / 10**9, mass, linestyle='dotted', color='black', label='M (M$_\odot$)')
plt.plot(ages / 10**9, luminosity, linestyle='dashed', color='black', label='log(L)')
plt.axvline(x = 4.9447584551452723, color='red')
plt.text(4.85, -2, '$t=4.94$ Gyr', fontsize = 10)

plt.xlim(4.7,5.2)
#xlim(0,10)
plt.xlabel("Age (Gyr)")

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()

#plt.savefig('Kippenhahn-1.1.jpeg', dpi=1200)
```

```python
#Kippenhahn-diagram-2

# make a MesaData object from a history file
h = mr.MesaData('history.data')

# extract the star_age column of data
ages = h.data('star_age')
log_R = h.data('log_R')
center_rho = h.data('log_center_Rho')
center_T = h.data('log_center_T')

plt.plot(ages / 10**9, log_R, linestyle='dotted', color='black', label='log(R)')
plt.plot(ages / 10**9, center_rho, linestyle='dashdot', color='black', label=r'$\rho_c$')
plt.plot(ages / 10**9, center_T, linestyle='solid', color='black', label='$T_c$')
plt.axvline(x = 4.9447584551452723, color='red')
plt.text(4.85, -1, '$t=4.94$ Gyr', fontsize = 10)

plt.xlim(4.7,5.2)
#xlim(0,10)
plt.xlabel("Age (Gyr)")

plt.legend()
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()

#plt.savefig('Kippenhahn-2.2.jpeg', dpi=1200)
```
